# Analyzing data to determine what car to buy

In this I would be scraping Kijiji listings that I have saved as favourites for cars that I am interested to help determine which one to buy.

```python
# import the necessary libraries

from bs4 import BeautifulSoup
from requests import get
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Scrape a sample auto page ad

Using beautiful soup, I would be scraping the url for one of the ads to get the basic structure of what I need before implementing it for the all the other pages using a loop.

```python
#get the kijiji auto ad page code
source = get('https://www.kijiji.ca/v-view-details.html?adId=1387072508').text

#parse the page code
soup = BeautifulSoup(source, 'lxml')
soup
```

```python
# Parse soup to grab attributes of the vehicle
attr = soup.find_all('dd', {'class':'attributeValue-2574930263'})

ad_name = soup.select_one("h1[class*=title-2323565163]").text
price = soup.find_all(
    'span', {'class':'currentPrice-2842943473'})[0].span.text.strip('$')
year = attr[1].text
make = attr[2].text
model = attr[3].text
trim = attr[4].text
colour = attr[5].text
body_type = attr[6].text
drivetrain = attr[7].text
fuel_type = attr[9].text
kilometers = attr[10].text
description = soup.find_all(
    'div', attrs={'class' : 'descriptionContainer-3544745383'})[0].text


# print out all the attributes
print('Here are the attributes of the ad and vehicle.\n\nAd Name: {}\nPrice: {}\nYear: {}\nMake: {}\nModel: {}\nTrim: {}\nColour: {}\nBody type: {}\nDrivetrain: {}\nFuel type: {}\nKilometers: {}\nDescription: \n{}'.format(ad_name, price, year, make, model, 
      trim, colour, body_type, drivetrain,
      fuel_type, kilometers, description))
```
